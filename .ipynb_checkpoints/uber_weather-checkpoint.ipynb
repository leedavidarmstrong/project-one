```python
import pandas as pd
```

```python
csv_file_path = 'Resource/uber.csv'
```

```python
uber_df = pd.read_csv(csv_file_path)
```

```python
#print(uber_df.head(4))
uber_df.head(4)
```

```python
#print(uber_df.tail(4))
uber_df.tail(4)
```

```python
middle_index = len(uber_df) // 2

#print(uber_df.iloc[middle_index - 2 : middle_index + 2])
uber_df.iloc[middle_index - 2 : middle_index + 2]
```

```python
# Sort the DataFrame by 'key' and ascending datetime
uber_df_ascending = uber_df.sort_values(by='key', ascending=True)

# Reset the index
uber_df_ascending = uber_df_ascending.reset_index(drop=True)

# Print the first four and midle four and last four rows of sorted DataFrame
print("\nFirst four rows:")
print(uber_df_ascending.head(4))

middle_index = len(uber_df_ascending) // 2

print("Middle four rows:")
print(uber_df_ascending.iloc[middle_index - 2 : middle_index + 2])

print("\nLast four rows:")
print(uber_df_ascending.tail(4))
```

```python
uber_df_ascending.to_csv('Resource/uber_df_ascending.csv', index=False)
```
