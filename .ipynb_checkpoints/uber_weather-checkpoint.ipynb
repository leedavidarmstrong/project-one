```python
# Import API key
from api_keys import geoapify_key
import pandas as pd
import hvplot.pandas

# Import API key
from api_keys import geoapify_key
```

```python
# Reading csv file and saving it as a dataframe
csv_file_path = 'Resource/uber.csv'
uber_df = pd.read_csv(csv_file_path)
```

```python
print(len(uber_df))
uber_df.head()
```

```python
# Sort the DataFrame by 'key' and ascending datetime
uber_df_sorted = uber_df.sort_values(by='key', ascending=True)

# Reset the index
uber_df_sorted = uber_df_sorted.reset_index(drop=True)
uber_df_sorted.head()
```

```python
# Configure the map plot
map_plot_1 = uber_df_sorted.hvplot.points(
    "pickup_longitude",
    "pickup_latitude",
    geo = True,
    tiles = "OSM",
    size = "fare_amount",
    scale = 1,
    color = "passenger_count"
)

# Display the map
map_plot_1
```
